```python
import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofnewyork.us", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofnewyork.us,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("k397-673e", limit = 4000000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```

```python
len(results_df)
```

```python
payroll_df = results_df.copy(deep = True)
```

```python
payroll_df[['fiscal_year']] = payroll_df[['fiscal_year']].astype(int)
payroll_df[['base_salary', 'regular_hours', 'regular_gross_paid', 
            'ot_hours', 'total_ot_paid','total_other_pay']] = payroll_df[['base_salary', 'regular_hours', 'regular_gross_paid', 
                                             'ot_hours', 'total_ot_paid', 'total_other_pay']].astype(float)
payroll_df['agency_start_date'] = payroll_df['agency_start_date'].str.replace("T00:00:00.000", "")
payroll_df['agency_start_date'] = pd.to_datetime(payroll_df['agency_start_date'], format = '%Y/%m/%d', errors = 'coerce')
payroll_df.head()
```

```python
payroll_trimmed_df = payroll_df[['fiscal_year', 'agency_name', 'last_name', 'first_name', 'mid_init', 'agency_start_date',
                                'work_location_borough', 'title_description', 'base_salary', 'pay_basis',
                                'regular_hours', 'regular_gross_paid', 'ot_hours', 'total_ot_paid', 'total_other_pay']]
payroll_trimmed_df.head()
```
